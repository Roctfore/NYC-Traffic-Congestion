#Build model

```python
import numpy as np
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#Obtain feature and target values
Y = data['SPEED'].values.reshape(-1,1)
X = data.drop(columns='SPEED').values
```

```python
#Standardized processing
SS_X = StandardScaler()
SS_Y = StandardScaler()
x = SS_X.fit_transform(X)
y = SS_Y.fit_transform(Y)
```

```python
#Divide the training and validation sets
x_train,x_test,y_train,y_test = train_test_split(x[::30],y[::30],random_state=42,shuffle=True,test_size=0.2)
```

```python
#Random forest regression
rf = RandomForestRegressor(n_estimators=100)
rf.fit(x_train, y_train)
```

```python
#Forecast
rf_y_pred = rf.predict(x_test)
```

```python
#Inverse normalization
rf_y_pred_invers = SS_Y.inverse_transform(rf_y_pred.reshape(-1,1))
y_test_invers = SS_Y.inverse_transform(y_test)
```

```python
#Output evaluation index
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('mse:',mean_squared_error(y_test_invers,rf_y_pred_invers))
print('mae:',mean_absolute_error(y_test_invers,rf_y_pred_invers))
print('r2:',r2_score(y_test_invers,rf_y_pred_invers))
```

```python
#Drawing to show the results
fig, ax = plt.subplots(dpi=300)
ax.plot(rf_y_pred_invers[::10], label='Predicted')
ax.plot(y_test_invers[::10], label='True')
ax.set_xlabel('Sample Index')
ax.set_ylabel('y Value')
ax.set_title('Random Forest Regression')
ax.legend()
plt.show()
```

```python
#Support vector regression model
svr = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr.fit(x_train, y_train)
```

```python
#Forecast
svr_y_pred = svr.predict(x_test)
```

```python
#Inverse normalizatio
svr_y_pred_invers = SS_Y.inverse_transform(svr_y_pred.reshape(-1,1))
```

```python
#Output evaluation index
print('mse:',mean_squared_error(y_test_invers,svr_y_pred_invers))
print('mae:',mean_absolute_error(y_test_invers,svr_y_pred_invers))
print('r2:',r2_score(y_test_invers,svr_y_pred_invers))
```

```python
#Drawings to show the results
fig, ax = plt.subplots(dpi=300)
ax.plot(svr_y_pred_invers[::10], label='Predicted')
ax.plot(y_test_invers[::10], label='True')
ax.set_xlabel('Sample Index')
ax.set_ylabel('y Value')
ax.set_title('SVR')
ax.legend()
plt.show()
```
