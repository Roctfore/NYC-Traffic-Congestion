## In order to simulate the interaction between districts in the city, we perform the network analysis on taxizones.

```python
Taxi=nx.Graph()
```

```python
Taxi.add_nodes_from(Taxizones.LocationID)
```

```python
links = pd.read_parquet('D:/fhv_tripdata_2022-06.parquet')
```

```python
links.dropna(subset=['PUlocationID'], inplace=True)
```

```python
links.dropna(subset=['DOlocationID'], inplace=True)
```

```python
links = links.reset_index()
```

```python
EN=len(links.index)
```

```python
edgelist=[(links['PUlocationID'][j],links['DOlocationID'][j]) for j in range(EN)]
```

```python
Taxi.add_edges_from(edgelist)
```

```python
components = list(nx.connected_components(Taxi))
```

```python
print(f"There are {len(components)} connected components in the network.")
for i, component in enumerate(components):
    size = len(component)
    print(f"Component {i+1}: {size} nodes")
```

```python
largest = max(nx.connected_components(Taxi), key=len)
Gc = Taxi.subgraph(largest)
```

```python
degree_centrality = nx.degree_centrality(Gc)
betweenness_centrality = nx.betweenness_centrality(Gc)
closeness_centrality = nx.closeness_centrality(Gc)
pagerank = nx.pagerank(Gc, alpha=0.85)
```

```python
top_10_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10]
top_10_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:10]
top_10_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:10]
top_10_pagerank = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)[:10]
```

```python
print("Top 10 in terms of degree centrality:")
for node, centrality in top_10_degree:
    print("Node: {}, Degree Centrality: {}".format(node, centrality))
```

```python
print("Top 10 in terms of betweenneass centrality:")
for node, centrality in top_10_betweenness:
    print("Node: {}, betweenneass Centrality: {}".format(node, centrality))
```

```python
print("Top 10 in terms of closeness centrality:")
for node, centrality in top_10_closeness:
    print("Node: {}, closeness Centrality: {}".format(node, centrality))
```

```python
print("Top 10 in terms of pagerank:")
for node, centrality in top_10_pagerank:
    print("Node: {}, pagerank: {}".format(node, centrality))
```

```python
shortest_paths = dict(nx.shortest_path_length(Gc))
```

```python
max_distance = max([max(lengths.values()) for lengths in shortest_paths.values()])
```

```python
print(f"The maximum shortest path distance between any two taxi zones is {max_distance}.")
```

```python
top_100_pagerank = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)[:100]
```

```python
top_airport_names = [Flights[0] for Flights in top_100_pagerank]
```

```python
total_distance = 0
n_pairs = 0
```

```python
for i in range(len(top_airport_names)):
    for j in range(i+1, len(top_airport_names)):
        try:
            distance = nx.shortest_path_length(Gc, source=top_airport_names[i], target=top_airport_names[j])
            total_distance += distance
            n_pairs += 1
        except nx.NetworkXNoPath:
            pass

avg_distance = total_distance / n_pairs
print(f"The average shortest path distance among the top 100 taxi zones by PageRank is {avg_distance:.2f}.")
```
